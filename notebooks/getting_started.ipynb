# Getting Started

`acs_nativity` provides three functions for accessing and analyzing historical data on the native and foreign-born population in the United States:

  * `get_nativity_timeseries`
  * `plot_nativity_timeseries`
  * `plot_nativity_change`

Data comes from the [American Community Survey (ACS) 1-Year Estimates](https://www.census.gov/data/developers/data-sets/acs-1year.html) and is available for: the nation, all states, the District of Columbia, Puerto Rico, all congressional districts and metropolitan statistical areas, and all counties and places (i.e., towns or cities) with populations of 65,000 or more.

## Example: Analyzing National Nativity Data

Below is a minimal example analyzing national-level nativity data:

```python
from acs_nativity import (
    get_nativity_timeseries,
    plot_nativity_timeseries,
    plot_nativity_change,
)

df = get_nativity_timeseries(us="*")
df.tail()
```

```python
plot_nativity_timeseries(
    df,
    column="Foreign-born",
    title="Foreign-Born Population in the United States<br>ACS 1-Year Estimates",
    y_label="Population",
)
```

```python
plot_nativity_change(
    df,
    "Foreign-born",
    "Change in the US Foreign-Born Population<br>ACS 1-Year Estimates",
)
```

## Working with Other Geographies

get_nativity_timeseries supports the same geography arguments used by the `censusdis` package. You specify a geography by passing keyword arguments such as `us`, `state`, `county`, or `place`.

The general pattern is:
  * Use `us="*"` for national data
  * Use `state=` for state-level data
  * Combine `state=` with additional keywords like `county=` or `place=` to drill down further

Below are concrete examples of these patterns.

```python
# For the entire country, use the keyword `us` and the value `*`
df = get_nativity_timeseries(us="*")

# For a state, use the keyword `state` and the value from `censusdis.states`
from censusdis.states import MN

df = get_nativity_timeseries(state=MN)

# County constants are organized by state
from censusdis.states import NY
from censusdis.counties.new_york import NASSAU

df = get_nativity_timeseries(state=NY, county=NASSAU)

# City constants are in the state-specific `places` module
from censusdis.states import IL
from censusdis.places.illinois import CHICAGO_CITY

df = get_nativity_timeseries(state=IL, place=CHICAGO_CITY)
```

For a full list of supported geographies and constants, see the [censusdis "Additional Geographies" documentation](https://censusdis.readthedocs.io/en/stable/intro.html#additional-geographies).
