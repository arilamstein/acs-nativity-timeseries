```python
from censusdis.multiyear import download_multiyear
from censusdis.datasets import ACS1
from plotting import plot_acs_timeseries

df = download_multiyear(
    dataset=ACS1,
    vintages=[year for year in range(2009, 2025) if year != 2020],
    group="B05012",
    us="*",
)
df
```

```python
plot_acs_timeseries(
    df,
    column="Foreign-born",
    title="Foreign-Born Population in the United States<br>ACS 1-Year Estimates",
    y_label="Population",
)
```
